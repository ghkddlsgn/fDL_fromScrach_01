```python
def AND(x1, x2):
    w1, w2, theta = 0.5, 0.5, 0.7
    tmp = w1 * x1 + w2 * x2
    return 1 if tmp > theta else 0
```

```python
import numpy as np

def OR(x:np.array):
    theta = 1
    w = [1, 1]
    b = 0

    y_hat = np.sum((w * x)+b)
    return 1 if y_hat >= theta else 0

print(OR(np.array([0,0])))
print(OR(np.array([1,0])))
print(OR(np.array([0,1])))
print(OR(np.array([1,1])))
```

```python
def step_function(x):
    y = x > 0
    return y.astype(int)
```

```python
def relu(x):
    return np.maximum(0,x)
```
